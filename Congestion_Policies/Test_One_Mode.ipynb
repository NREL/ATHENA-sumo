# Test 1: 100% to one mode
For this test we are going to allocate 100% of the vehicles to one mode and make all the people be 50% Dropoff and 50% pick-up(Departing) patrons. 

We anticipate the following:
- All vehicles using one mode
- No extra Buses added
- Pass thru vehicles will be conserved 

```python
num_modes = 7   # Number of modes  
mode_strings = {0:'_passenger_',1:'_TNC_',2:'_Taxi_', 3:'Transit',
                4:'_parking_',5:'_Limo_', 6:'Rental', 7:'pass_'}
curb_trips = (demand['A'] + demand['B'] + demand['C'] + demand['D'] + demand['E'] + demand['parking']).values[0]
all_trips = (demand['A'] + demand['B'] + demand['C'] + demand['D'] + demand['E'] + demand['parking'] + demand['pass']).values[0]
```

```python
for mode_index in range(7):
    mode_percents = np.zeros(num_modes, dtype=int) # array used to sent the mode percentages, assumes index 0 refer to personal, 
    # 1 to TNC, 2 to Taxi, 3 to Transit, 4 to parking, 5 to limo and 6 to rental
    mode_percents[mode_index] = 1
    
    print("*"*90)
    print("Testing for Mode: ", mode_strings[mode_index])
    
    master_function(demand=demand,
                dwell_times=dwell_time,
                Date='2017-10-10',
                percentOfPassenger=float(mode_percents[0]), 
                percentDropOff=0.5,
                percentPickUp=0.5,
                percentOfTNC=float(mode_percents[1]),
                percentOfTaxi=float(mode_percents[2]),
                percentOfTransit= float(mode_percents[3]),
                percentOfParking= float(mode_percents[4]),
                percentOfLimo= float(mode_percents[5]),
                percentOfRental= float(mode_percents[6]),
                Delete_Files = True,
               level='TEST')
    
    #Name of trip file that was generated
    trip_file = '../Example_Files/TempInputTrips/2017-10-10.TEST.trips.xml'

    #calling function
    conservation = 0
    count = 0

    for index, mode in mode_strings.items():
        print("Finding number of trips for mode ")
        num_trip_mode = num_trips_by_mode(trip_file, mode)

        if mode =="Rental" or mode =="Transit":
            assert num_trip_mode == 0   # rental or transit do not have trips in the final trip file
            print(u'\u2713',mode, " trips is zero")

        elif mode == 'pass_':
            assert num_trip_mode == demand['pass'].values[0]
            print(u'\u2713',"Pass through trips are conserved")
        else: 
            assert num_trip_mode == curb_trips * int(mode_percents[index])
            print(u'\u2713', mode, " is conserved")

        conservation += num_trip_mode

    if mode_index != 3 and mode_index != 6: 
        print("mode is ", mode_index)
        assert conservation == all_trips
        print(u'\u2713',"Conservation is conserved")
```
